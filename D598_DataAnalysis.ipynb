## Filtering data to return businesses with negative debt to equity ratios

## Calculating debt to income ratios

```python
# Calculate Debt-to-Income Ratio
df['Debt to Income Ratio'] = df['Total Long-term Debt'] / df['Total Revenue']
```

```python
df
```

<p>These are the debt to income ratio for all businesses. Debt to income ratio was calculated by dividing total long-term debt by total revenue. An additional column was added to the main data frame.</p>

## Creating a new data frame that only includes the debt to income ratios for all businesses

```python
df2 = df[['Business ID','Debt to Income Ratio']]
df2
```

<p> This is a new data frame that only contains two columns: business ID and debt to income ratio. The column, debt to income ratio, is a newly created column from the previous step. This data frame will be merged with the original data frame in the next step.</p>

## Concatenating debt-to-income ratio data frame 'df2' with the original data frame 'df'

```python
concatenated_df = pd.concat([df.drop(columns='Debt to Income Ratio'), df2['Debt to Income Ratio']], axis=1)
```

```python
concatenated_df
```

<p>Here we concatenated the newly created data frame 'df2' that contains the debt to income ratio for businesses with the original data frame 'df.' The final data frame now contains the debt to income ratio. We had to drop the debt to income ratio column in the original data frame to show that we can merge two different data frames using the concat function.</p>

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Visualizations

```python
plt.figure(figsize=(10, 6))

sns.regplot(x=df['Total Liabilities'], y=df['Total Revenue'], scatter=True, color='orange', line_kws={"color": "orange"})
plt.title('Total Liabilities vs Total Revenue')
plt.xlabel('Total Liabilities')
plt.ylabel('Total Revenue')
plt.grid(True)

plt.show()
```

```python
sort_df = df.sort_values(by='Total Revenue', ascending=False)
plt.figure(figsize=(10, 6))

plt.bar(sort_df['Business State'], sort_df['Total Revenue'], color='orange')
plt.title('Total Revenue by Business State')
plt.xlabel('Business State')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90, fontsize=10) 
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.hist(df['Debt to Income Ratio'], bins=4, color='orange', alpha=0.8)
plt.title('Distribution of Debt to Income Ratio')
plt.xlabel('Debt to Income Ratio')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.8)

plt.show()
```

```python
grouped_data = df.groupby('Business State')['Total Liabilities'].sum()
top_5 = grouped_data.sort_values(ascending=False).head(5)
plt.figure(figsize=(8, 8))

plt.pie(top_5, labels=top_5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 States by Total Liabilities')
plt.show()
```
